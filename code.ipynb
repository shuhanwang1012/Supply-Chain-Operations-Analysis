```python
# Create a pie chart of the 'column_name' column
column_counts = df['Department Owner/Investigator'].value_counts()

plt.figure(figsize=(10, 10))
pie_chart = column_counts.plot(kind='pie', autopct='%.1f%%', pctdistance=0.8)

# Set the legend to be outside the pie chart and add a title
pie_chart.legend(loc='center left', bbox_to_anchor=(1, 0.5))
pie_chart.set_title('Department Owner/Investigator')
```
